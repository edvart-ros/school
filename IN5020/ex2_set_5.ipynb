```python
import numpy as np
```

```python
X = [1.0,2.9,4.2,5.1,7.3,5.4,7.9,8.8,9.8,11.2]
sigma = 2

def P(x_k, mu_1_hat, mu_2_hat, mu_i_hat):
    num = np.exp((-1/2)*((x_k-mu_i_hat)/sigma)**2)
    denom = np.exp((-1/2)*((x_k-mu_1_hat)/sigma)**2) + np.exp((-1/2)*((x_k-mu_2_hat)/sigma)**2)
    return num/denom

def get_next_mu_1(mu_1, mu_2):
    num = 0
    denom = 0
    for x_k in X:
        num += P(x_k, mu_1, mu_2, mu_1)*x_k
        denom += P(x_k, mu_1, mu_2, mu_1)
    return num/denom

def get_next_mu_2(mu_1, mu_2):
    num = 0
    denom = 0
    for x_k in X:
        num += P(x_k, mu_1, mu_2, mu_2)*x_k
        denom += P(x_k, mu_1, mu_2, mu_2)
    return num/denom
```

```python
mu_1 = 2
mu_2 = 10


for i in range(1000):
    mu_1 = get_next_mu_1(mu_1, mu_2)
    mu_2 = get_next_mu_2(mu_1, mu_2)
    print(mu_1, mu_2)
```
